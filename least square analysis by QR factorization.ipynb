```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
print("Question 40. \n The solution is \n")
print("By applying Least Square Analysis using QR factorization: ")
print("Ax=b and A=QR hence By substituting, we get \n X=(R^-1).Q'.b \n")
#enter your matrix here
A1=np.array([[1,8,64],[1,9,81],[1,10,100],[1,11,121],[1,12,144],[1,13,169]])
A2=np.array(([1,8,64,512],[1,9,81,729],[1,10,100,1000],[1,11,121,1331],[1,12,144,1728],[1,13,169,2197]))
b =np.array([[29.3],[32.0],[32.5],[32.7],[31.7],[31.2]])
q1, r1 = np.linalg.qr(A1)
q2, r2 = np.linalg.qr(A2)
print("The q matrix for Quadratic Model")
print(q1)
print("The r matrix for Quadratic Model")
print(r1)
print("The q matrix for Cubic Model")
print(q2)
print("The r matrix for Cubic Model")
print(r2)
x1=(np.linalg.inv(r1)).dot((q1.transpose()).dot(b))
x2=(np.linalg.inv(r2)).dot((q2.transpose()).dot(b))

e = np.linspace(8, 13, 6)
y1 = x1[0] + x1[1] * e + x1[2] * (e ** 2)
y2 = x2[0] + x2[1] * e + x2[2] * (e ** 2) + x2[3] * (e ** 3)
plt.plot(e, y1)
plt.plot(e, y2)
plt.show()

p = mean_squared_error(b, y1)
q = mean_squared_error(b, y2)
print("The MSE for Quadratic model is {0}".format(p))
print("The MSE for Cubic Model is {0}".format(q))

print("Since the MSE for the Quadratic equation is less than that of the cubic, it will give us better "
          "predictions")
u=np.linspace(10,25,10)
v=np.linspace(0,15,15)
y3=x2[0]+x2[1]*u+x2[2]*(u**2)+x2[3]*(u**3)
y4=x2[0]+x2[1]*v+x2[2]*(v**2)+x2[3]*(v**3)
plt.plot(u,y3,'o')
plt.plot(v,y4)
plt.show()
#Below are the predictions for future years
print("The predicted values are:")
print(y3)
```
